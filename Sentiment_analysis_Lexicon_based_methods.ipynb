```python
import pandas as pd
import re
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER sentiment lexicon
nltk.download('vader_lexicon')
```

```python
def sentiment_scores(sentence):
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)

    print("Sentence sentiment analysis:", sentiment_dict)
    print(sentiment_dict['neg'] * 100, "% Negative")
    print(sentiment_dict['neu'] * 100, "% Neutral")
    print(sentiment_dict['pos'] * 100, "% Positive")

    print("Sentence Overall Rated As", end=" ")
    if sentiment_dict['compound'] >= 0.05:
        print("Positive")
        label = "Positive"
    elif sentiment_dict['compound'] <= -0.05:
        print("Negative")
        label = "Negative"
    else:
        print("Neutral")
        label = "Neutral"

    return label, sentiment_dict['compound']
```

```python
# Load the CSV file
df = pd.read_csv("tweets-data.csv")

# Take a sample of 500 rows
df_sample = df.sample(n=500, random_state=42).reset_index(drop=True)
```

```python
def clean_text(text):
    text = str(text)
    text = re.sub(r"http\S+", "", text)          # Remove URLs
    text = re.sub(r"@\w+", "", text)             # Remove mentions
    text = re.sub(r"#", "", text)                # Remove hashtag symbols
    text = re.sub(r"[^A-Za-z\s]", "", text)      # Keep only letters
    text = text.lower().strip()                  # Convert to lowercase
    return text

# Apply cleaning
df_sample['cleaned_text'] = df_sample['Tweets'].apply(clean_text)
```

```python
# Apply the sentiment function to get label and score
df_sample[['sentiment_label', 'sentiment_score']] = df_sample['cleaned_text'].apply(lambda x: pd.Series(sentiment_scores(x)))
```
